```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("dataset.csv", sep=",")
df
```

```python
df.dtypes # Se listan los tipos de datos en las columnas
```

```python
df['date'] = pd.to_datetime(df['timestamp']) # Se conviente la columna timestamp a fecha (date)
df.pop('timestamp')#Se borra la columna timestamp
```

```python
# Se van a crear 2 columnas con el aÃ±o y el mes
df['year'] = pd.DatetimeIndex(df['date']).year
df['month'] = pd.DatetimeIndex(df['date']).month
df
```

```python
df['pollutant'].unique() # Se analizan que tipos de contaminantes que tienen el dataset
```

```python
df = df.sort_values(by=['year','month'])#ordena por aÃ±o y mes
```

```python
df_aux = df.loc[:,['city','country','year','month','pollutant','value','unit']] #Se crea otro dataset solo con las columnas que me interesan
df_aux = df_aux.loc[(df_aux['value']>0),:]# Se filtran solo mediciones mayores a 0
df_aux = df_aux.fillna(0) # Se rellenan los valores nulos con 0
```

```python
#pd.set_option('display.max_rows', 500)
pd.set_option('display.max_rows', 10) # Establece cuantas filas se pueden mostrar, CUIDADO PORQUE SI EL DATASET TIENE MUCHAS FILAS PUEDE TRANCAR EL JUPYTER
df_means_pollutant_by_month = df_aux.groupby(['year','month','pollutant']).mean() # Muestra la media de los valores agrupados por aÃ±o y mes
df_means_pollutant_by_month
```

```python
# Se crea una tabla pivote con los aÃ±os como indices y los componentes como columnas
pv = pd.pivot_table(df_aux, index=df_aux.year, columns=df_aux.pollutant,
                    values='value', aggfunc='mean')
pv.fillna(0) # Donde no tiene valores se los rellena con 0
```

```python
# Como no se puede apreciar correctamente porque los valores ya que tienen diferentes escalas, se pueden separar por contaminante
pv.plot(subplots=True, figsize=(10, 10));
```
